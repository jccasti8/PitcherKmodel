```python
def get_pitcher_id(first, last):
    return int(playerid_lookup(last, first).iloc[0]['key_mlbam'])

cole_id = get_pitcher_id('gerrit', 'cole')
cole_data = statcast_pitcher('2023-04-01', '2023-09-30', cole_id)
cole_data.head()
               
```

```python
pitcher_team = 'NYY'

strikeouts = cole_data[cole_data['events'].isin(['strikeout', 'strikeout_double_play'])]
k_per_game = strikeouts.groupby('game_date').size().reset_index(name='strikeouts')

# Step 2: Get home/away info and derive opponent
game_info = cole_data.groupby('game_date').agg({
    'home_team': 'first',
    'away_team': 'first'
}).reset_index()

def get_opponent(row):
    return row['away_team'] if row['home_team'] == pitcher_team else row['home_team']

game_info['opponent'] = game_info.apply(get_opponent, axis=1)

# Step 3: Merge into final dataset
k_per_game = k_per_game.merge(game_info[['game_date', 'opponent']], on='game_date')
k_per_game['pitcher'] = 'Gerrit Cole'

k_per_game.head()
```

```python
k_per_game = k_per_game.merge(game_info[['game_date', 'home_team']], on='game_date')
k_per_game['home'] = k_per_game['home_team'] == 'NYY'
k_per_game.drop(columns='home_team', inplace=True)
```

```python
#Add days of rest 
k_per_game['game_date'] = pd.to_datetime(k_per_game['game_date'])

k_per_game.sort_values('game_date', inplace=True)

k_per_game['days_rest'] = k_per_game['game_date'].diff().dt.days
```

```python
#Add rolling K avg (recent form)
k_per_game['k_rolling_avg_3'] = k_per_game['strikeouts'].rolling(window=3).mean().shift(1)
```

```python
# Quick mapping — can expand as needed
team_map = {
    'SF': 'SFG', 'PHI': 'PHI', 'BAL': 'BAL', 'CLE': 'CLE', 'MIN': 'MIN', 'TB': 'TBR',
    'TOR': 'TOR', 'HOU': 'HOU', 'BOS': 'BOS', 'KC': 'KCR', 'CIN': 'CIN',
    'SEA': 'SEA', 'SD': 'SDP', 'OAK': 'OAK', 'LAD': 'LAD', 'MIL': 'MIL',
    'CHC': 'CHC', 'TEX': 'TEX', 'DET': 'DET', 'MIA': 'MIA', 'WSH': 'WSN',
    'NYM': 'NYM', 'ATL': 'ATL', 'STL': 'STL', 'PIT': 'PIT', 'ARI': 'ARI',
    'COL': 'COL', 'CHW': 'CHW', 'LAA': 'LAA'
}

k_per_game['opponent_fg'] = k_per_game['opponent'].map(team_map)
```

```python
# Select only the columns we care about
team_k_rates = team_batting_2023[['Team', 'K%']].copy()
team_k_rates.rename(columns={'K%': 'team_k_rate'}, inplace=True)

# Merge with our pitcher data
k_per_game = k_per_game.merge(
    team_k_rates,
    left_on=k_per_game['opponent'].map(team_map),
    right_on='Team',
    how='left'
)

# Clean up
k_per_game.drop(columns='Team', inplace=True)
```
